```python
# In this notebook, I will be using Scikit-Learn's KMeans class to perform Color Segmentation on images.
```

```python
import numpy as np
import sklearn
from sklearn.cluster import KMeans
import matplotlib
from matplotlib.image import imread
import matplotlib.pyplot as plt
```

```python
image = imread('lena.png')
image.shape    
```

```python
image_flattened = image.reshape(-1,image.shape[2])
image_flattened.shape
```

```python
clusters = [64,32,16,8,4,2]
plt.imshow(image)
plt.title('Original Image')
plt.show()
for c in clusters:
    kmeans = KMeans(n_clusters = c)
    kmeans.fit(image_flattened)
    segmented_img_flattened = kmeans.cluster_centers_[kmeans.labels_]
    segmented_img = segmented_img_flattened.reshape(image.shape)
    plt.imshow(segmented_img)
    plt.title('Color Segmented Image with {} colors'.format(c))
    plt.show()
```
