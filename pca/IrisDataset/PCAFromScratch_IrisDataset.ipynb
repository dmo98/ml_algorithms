```python
# Import Statements
import numpy as np
import sklearn
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
# Load the data
iris_data = load_iris()
iris_data.keys()
```

```python
# Separate the predictors and the target  variable
X = iris_data['data']
Y = iris_data['target']

print(X.shape, Y.shape)
```

```python
def CenterData(array):
    '''This function centers the data to the origin.'''
    mu = np.zeros((array.shape[1],1))
    
    for j in range(array.shape[1]):
        mu[j] = np.mean(array[:,j])
        
        array[:,j] = array[:,j] - mu[j]
    return array, mu

def PrincipalComponents(array):
    '''This function returns the principal components of the dataset.'''
    U, s, Vt = np.linalg.svd(array)
    return Vt

def ProjectData(array, pc, d):
    '''This function projects the data onto the hyperplane defined by the first "d" principal components of the data.'''
    X_projected = array.dot(pc.T[:,:d])
    return X_projected

#X_centered, mu = CenterData(X)
#print('Computed Feature Means: ', mu.flatten())

#Vt = PrincipalComponents(X_centered)
#print('Principle Components matrix: ', '\n', Vt.T)

#X_2d = ProjectData(X_centered, Vt, 2)
#X_3d = ProjectData(X_centered, Vt, 3)
```

```python
def PCA(X, d):
    '''This function runs PCA on the data and returns the projected data.'''
    X_centered, mu = CenterData(X)
    Vt = PrincipalComponents(X_centered)
    X_projected = ProjectData(X_centered, Vt, d)
    return X_projected

X_2d = PCA(X,2)
X_3d = PCA(X,3)
```

```python
plt.scatter(X_2d[:,0], X_2d[:,1], c = Y)
plt.title('Iris Dataset projected down to 2D',fontsize=20,fontstyle='italic')
plt.savefig('IrisDatasetProjectedDownTo2D.png', dpi = 300)
plt.show()
```

```python
plt.figure(figsize = (8, 8))
ax = plt.axes(projection ="3d")
ax.scatter3D(X_3d[:,0], X_3d[:,1], X_3d[:,2], c = Y)
plt.title('Iris Dataset projected down to 3D',fontsize=20,fontstyle='italic')
plt.savefig('IrisDatasetProjectedDownTo3D.png', dpi = 300)
plt.show()
```
