```python
# In this notebook, I will be using Scikit-Learn's PCA class to perform dimensionality reduction on the Iris Dataset for 
# DataViz
```

```python
import numpy as np
import sklearn
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
iris_data = load_iris()
iris_data.keys()
```

```python
X = iris_data['data']
Y = iris_data['target']

print(X.shape, Y.shape)
```

```python
from sklearn.decomposition import PCA

pca_2d = PCA(n_components = 2)
pca_3d = PCA(n_components = 3)

X_2d = pca_2d.fit_transform(X)
X_3d = pca_3d.fit_transform(X)
```

```python
print('Explained Variance Ratio of the PCA that projects X down to ')
print('\t 2D: {}'.format(pca_2d.explained_variance_ratio_))
print('\t 3D: {}'.format(pca_3d.explained_variance_ratio_))
```

```python
plt.scatter(X_2d[:,0], X_2d[:,1], c = Y)
plt.title('Iris Dataset projected down to 2D',fontsize=20,fontstyle='italic')
#plt.savefig('IrisDatasetProjectedDownTo2D.png', dpi = 300)
plt.show()
```

```python
plt.figure(figsize = (8, 8))
ax = plt.axes(projection ="3d")
ax.scatter3D(X_3d[:,0], X_3d[:,1], X_3d[:,2], c = Y)
plt.title('Iris Dataset projected down to 3D',fontsize=20,fontstyle='italic')
#plt.savefig('IrisDatasetProjectedDownTo3D.png', dpi = 300)
plt.show()
```
